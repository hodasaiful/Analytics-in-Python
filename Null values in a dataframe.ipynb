```python
import pandas as pd
import numpy as np
EPL = {
'NAME':['Salah','Christiano','Sergio','Virgil Van','Kevin De','Sadio Mane','Angelino','Jordan H']
,'CLUB':['Liverpool','Juventus','Man C','Liverpool','Man C','Liverpool','Man C','Liverpool']
,'AGE':[28,35,32,27,28,28,23,30]
,'SAL':[370000,np.nan, 320000,np.nan, 620000, np.nan,420000 ,270000 ]
,'GOALS':[42,55,27,32,10,40,9,37]
,'COUNTRY':['Egypt','Potugal','Argentina','Netherland','UK','Senegal','Spain','UK']
,'CONTRACT_ID':[10245,30465,56432,78904,10098,30243,56784,45009]   
}
EPLREC = pd.DataFrame(EPL)
```

```python
#NULL values in a specific column

EPLREC['SAL'].isnull().sum()
```

```python
#Drop any row with a Null Value

EPLREC.dropna(axis=0, inplace=True)
EPLREC
```

```python
#Drop any column with a Null Value

EPLREC.dropna(axis=1, inplace=True)
EPLREC
```

```python
#Drop missing values 

EPLREC.dropna(subset=['SAL'], axis=0, inplace=True)
```

```python
EPLREC
```

```python
#In this case, I populate the missing value with the mean of elements in the Salary column.
#You could write your own custom formula to poluate the missing values
EPLREC.fillna(value=EPLREC['SAL'].mean(), inplace=True)

#Print out the Dataframe
EPLREC
```
