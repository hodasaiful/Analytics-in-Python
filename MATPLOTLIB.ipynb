```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

### BAR CHART

```python
data = tips.groupby('day').sum()['total_bill']
plt.bar(data.index,data);
```

##### Bar Chart - Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

```python
#data as above
data = tips.groupby('day').sum()['total_bill']

#plot the data
plt.bar(data.index,data, width= 0.5, align='center', edgecolor='Orange',color='cyan');

# Adding and formatting title
plt.title("Day Versus Total Bill\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Day", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Total_Bill", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'});


plt.show()
```

### Pie Chart

```python
#Alter the default figure size
plt.figure(figsize=(10,6))

#Data Labels 
Employee = ['A', 'B', 'C','D', 'E']

#Data
Salary = [40000, 50000, 70000, 54000, 44000]

#Color mapped to each value
colors = ['#FF0000', '#0000FF', '#FFFF00','#ADFF2F', '#FFA500']

# explosion is used to separate out each slice of the doughnut (the white space in between)
explode = (0.0, 0.0, 0.0, 0.05, 0.0)

# Plot the Pie chart
plt.pie(Salary, colors=colors, labels=Employee,autopct='%1.1f%%', pctdistance=0.85,explode=explode)
#autopct is used to calculate the prcentage of each slice
#pctdistance is used to calculate the distance of autopct values from the plot

# Adding Title of chart
plt.title('Pie chart Demo\n', fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'});
```

### Transform a pie chart to a doughnut chart

- Build a pie chart
- add a smaller circle at the centre to mask the middle portion of the pie chart

```python
#Alter the default figure size
plt.figure(figsize=(10,6))

#Data Labels 
Employee = ['A', 'B', 'C','D', 'E']

#Data
Salary = [40000, 50000, 70000, 54000, 44000]

#Color mapped to each value
colors = ['#FF0000', '#0000FF', '#FFFF00','#ADFF2F', '#FFA500']

# explosion is used to separate out each slice of the doughnut (the white space in between)
explode = (0.03, 0.03, 0.03, 0.03, 0.03)

# Plot the Pie chart
plt.pie(Salary, colors=colors, labels=Employee,autopct='%1.1f%%', pctdistance=0.85,explode=explode)
#autopct is used to calculate the prcentage of each slice
#pctdistance is used to calculate the distance of autopct values from the plot


# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()

# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)

# Adding Title of chart
plt.title('Doughnut chart Demo\n', fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Brown'});

# Displaing Chart
plt.show()
```

# Sub Plots in Matplotlib

```python
#create 2 rows and 4 columns i.e a total of 2*4  8 plots
#share the x and Y axis


fig,a =  plt.subplots(nrows =2, ncols=4, sharey=True, sharex=True)
fig.suptitle("The main heading", fontsize=20)

import numpy as np
x = np.arange(1,5)
a[0][0].plot(x,x*x, color='maroon', label='legend',lw=2, alpha=0.5, linestyle ='--')
a[0][0].set_title('square', fontsize=18)
a[0][0].legend(loc='best')






#set the size
fig.set_size_inches(20.5, 10.5, forward=True)
```

#### Bringing together matplotlib and Seaborn

```python
#imports
import seaborn as sns
import matplotlib.pyplot as plt

#data 
fp = sns.load_dataset('flights')
flights = fp.pivot_table(index='month', columns ='year', values ='passengers')

#the seaborn part of the visual
sns.heatmap(flights
            , cmap ='BuPu'
            , linecolor ='white'
            ,linewidth =0.2
           ,);

#the matplotlib styling
plt.title("\nFlights by Month and Year\n", fontsize=20);
plt.xlabel("\n Year \n", fontdict={'fontsize': 18, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("\n Month \n", fontdict={'fontsize': 18, 'fontweight' : 5, 'color' : 'Brown'})

#change the styling of ticks
plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12);
```
