### Read from a csv file stored on the  machine at a specific folder

```python
#import pandas library with an alias of pd
import pandas as pd

#sepcify the path and file name on your local system; in the sysntx below
data_csv = pd.read_csv('C:\PYTHON\PYTHONCODE\GM.csv')

#check the contents (first 5 lines) of the dataframe
data_csv.head()
```

```python
#CSV data can be read using read_table, and pass an additional argument  
data_sep = pd.read_table('C:\PYTHON\PYTHONCODE\GM.csv', sep =',')

#check the contents (first 5 lines) of the dataframe
data_sep.head()
```

### Read in and display the files to standard output 

```python
import pandas as pd
data = pd.read_csv('C:\PYTHON\PYTHONCODE\corr.csv')
```

```python
'''
Display the contents to the standard output, and use a '|' 
as the separator
'''
import sys
data.to_csv(sys.stdout, sep ='|')
```

### Export a dataframe to an excel file on the local machine

```python
#excel_out.xlsx file gets created if not there; and data(dataframe) gets exported to it.

data.to_excel('C:\PYTHON\PYTHONCODE\excl_out.xlsx')
```

```python
#Missing values appear as empty string in the o/p.In order to populate with some other desired value 

data.to_excel('C:\PYTHON\PYTHONCODE\excl_out.xlsx', na_rep='NULL')
```

```python
#To remove both the columns and row header; disable index, header by setting the value to False.

data.to_excel('C:\PYTHON\PYTHONCODE\excl_out.xlsx', na_rep='NULL',index =False, header=False)
```

### Export a dataframe to a csv file on the local machine
